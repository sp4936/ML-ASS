```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

```python
Dataset=pd.read_csv('Mall_Customers.csv')
```

```python
Dataset
```

```python
X=Dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(
        n_clusters=i,
        init='k-means++',
        max_iter=300,
        random_state=0,
        algorithm='elkan'
    )
    kmeans.fit_predict(X)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(15,5))
plt.plot(range(1,11),wcss,marker='o',linestyle='-',color='blue')
plt.title('Elbow Curve')
plt.xlabel('cluster')
plt.ylabel('Wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',max_iter=300,random_state=0,algorithm='elkan')
ymeans=kmeans.fit_predict(X)
```

```python
plt.scatter(X[ymeans==0,0],X[ymeans==0,1],s=100,color='red',label='Cluster1')
plt.scatter(X[ymeans==1,0],X[ymeans==1,1],s=100,color='blue',label='Cluster1')
plt.scatter(X[ymeans==2,0],X[ymeans==2,1],s=100,color='pink',label='Cluster1')
plt.scatter(X[ymeans==3,0],X[ymeans==3,1],s=100,color='green',label='Cluster1')
plt.scatter(X[ymeans==4,0],X[ymeans==4,1],s=100,color='yellow',label='Cluster1')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,color='black')
plt.title('Kmeans Clusturing ')
plt.xlabel('Income')
plt.ylabel('Spending score')
plt.show()
```
